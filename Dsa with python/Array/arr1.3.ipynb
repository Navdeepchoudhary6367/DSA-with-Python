```python
# Find the max element

arr2 = arr.array ('i',[1,2,3,9,5])
print("Max value", max(arr2))

print("Min value", min(arr2))
```

```python
arr3 = arr.array ('i',[1,2,3,4,5])
odd = 0
even = 0

for i in arr3:
    if i % 2 ==0:
        even += 1
    else:
        odd += 1

print("Even no =",even)     
print("Odd no =",odd)            
```

```python
reversed_array = arr3[::-1]
print(reversed_array)
```

```python
# Search for an element

arr4 = arr.array ('i',[1,2,3,4,5,9,7,6])
x = int(input("Enter the no"))

if x in arr4:
    print(x,"index no", arr4.index(x))
else:
    print(x,"not found")

```

```python
avg= sum(arr4) / len(arr4)
print("Average",avg)
```
