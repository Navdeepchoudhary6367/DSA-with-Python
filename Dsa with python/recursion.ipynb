```python
# check if a number is palindrome or not

def is_palindrome(num):
    n = num
    result = 0

    while n>0:
        lastdigit = n % 10
        result = (result * 10) + lastdigit
        n = n//10


    return num == result


print(is_palindrome(3425))
print(is_palindrome(1221))

# Time complexity = O(log10(n))
# space complexity = O(1)
```

```python
#armstrong no.

def arm_strong(n):
    num = n
    total = 0
    nod = len(str(n))
    while num>0:
        ld = num % 10
        total = total + (ld**nod)
        num = num//10

    return total == n
    

print(arm_strong(153))
print(arm_strong(370))
print( arm_strong(345))



# time complexity  = O(log10(n))
# space complexity  = O(1)
```

```python
# print all factors

def all_factors(num):
    total = []
    for i in range(1 , num+1):
        if num % i == 0:
            total.append(i)
    return total

print(all_factors(20))  


# Time complexity = O(n)
# Space complexity = O(k )





def all_factor(n):
    total =[]
    for i in range (1, n//2 + 1):
        if n % i == 0:
            total.append(i)
    total.append(n)
    return total


print(all_factor(20))



        
    
```

```python
# store the frequency in dictionary

nums = [5,6,3,7,4,5,7,9,1,2,3,6,8]

freq = {}
for i in range (0, len(nums)):
    if nums[i] in freq:
        freq[nums[i]] += 1
    else:
        freq[nums[i]]  = 1    


print(freq)        
```

```python
# factorial using recursion

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
    

print(factorial(6))
```

```python
#fibonacci series using recursion

def fibo(n):
    if n <= 1:
        return n
    else:
        return fibo(n - 1) + fibo(n - 2)   

for i in range(6):
    print(fibo(i)) 
```

```python
# sum of first n numbers


def sum(n):
    if n == 0:
        return 0
    else:
        return n + sum(n-1)
    

print(sum(11))
```

```python
# reverse a array

def reverse(n, left, right):
    if left >= right:
        return
    n[left], n[right]= n[right] , n[left]
    reverse(n,left + 1, right - 1)

n = [3,4,5,3,7,8,9,1]
reverse(n,0,len(n)- 1)
print("reversed  array", n)

  

```
