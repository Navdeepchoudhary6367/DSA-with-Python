```python
b = arr.array('i',[1,2,3,4,5,6,9])
a.insert(0,4)   # this is used to add element at particular place
print(a)

c = arr.array('i',[1,2,3,4,5,6,4])
a.append(7)    # this is used to add element at end of array
print(a)

d = arr.array('i',[1,2,3,4,5,6,3])
a.extend([8,9])  # this is used to add multiple element at end of array
print(a)
```

```python
e = arr.array('i',[1,2,3,4,5,6,9,3,0])
print(e.pop())   # pop the last element and in pop we can return the element
print(e.pop(2))    # pop the specific element which index is given

f = arr.array('i',[1,2,3,4,5,7,8,6,9])
f.remove(7)
print(f)
```

```python
# adding two array

g = arr.array('i',[1,2,3,4,5,7,8,6,9])
h = arr.array('i',[5,7,8,6,9])
i = arr.array('i')
i = g+h
print(i)
```

```python
# Slicing

j = arr.array('i',[1,2,3,4,5,7,8,6,9])
print(j[0:4])
```

```python
print(j[::-1])
```

```python
for value in j:
    if value % 2 == 0 :
        print("It is divided")
    else:
        print("Not divided")   
```
