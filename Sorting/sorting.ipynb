```python
#Count the Number of Swaps
def selection_sort(n):
    count = 0

    for i in range(0,len(n)):
        min_index = i

        for j in range(i+1,len(n)):
            if n[j] < n[min_index]:
                min_index = j

        if min_index != i:
            n[i],n[min_index] = n[min_index], n[i]
            count += 1
    
    print("total no of swaps:" ,count)
    return n


numbers = [64, 25, 12, 22, 11, 13]
print(selection_sort(numbers))


# Time_complexity = O(n2)
# Space_complexity = O(1)
```

```python
# Bubble sort 


def bubble_sort(nums):
    n = len(nums)
    for i in range (n-2, -1, -1):

        for j in range(0,i+1):
            if nums[j] > nums[j+1]:
                nums[j], nums[j+1] = nums[j+1], nums[j]

    return nums


numbers = [3,4,2,6,8,55,4,22]
print(bubble_sort(numbers))



# Time_complexity = O(n2)
# Space_complexity = O(1)
```

```python
# Insertion sort

def insertion_sort(nums):
    n = len(nums)
    for i in range (1, n):
        key = nums[i]
        j = i-1

        while j >= 0 and nums[j] > key:
            nums[j+1] = nums[j]
            j -= 1

        nums[j+1] = key
    
    return nums

numbers = [3,4,2,6,8,55,4,22]
print(insertion_sort(numbers))


           
```

```python
# Merge Sort 


def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left_arr = arr[ :mid]
    right_arr = arr[mid: ]

    left = merge_sort(left_arr)
    right = merge_sort(right_arr)
    
    return merge_array (left,right)

def merge_array(left,right):
    result = []
    i,j = 0,0
    n,m = len(left), len(right)

    # Merge both halves
    while i<n and j<m:
        if left[i] <= right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1

    # Add remaining elements (if any)
    
    while i<n:
            result.append(left[i])
            i+=1

    
    while j<m:
            result.append(right[j])
            j+=1


    return result        


number = [27,31,52,76,23,34,67,33,66]
print(merge_sort(number))
```

```python
# Quick sort

def partition(nums,low,high):
    pivot = nums[low]
    i = low
    j = high 
    while i< j:
        while nums[i]<= pivot and i <= high-1:
            i += 1
        while nums[j]> pivot and j >= low-1:
            j-=1
        if i<j :
            nums[i], nums[j] = nums[j], nums[i]
    nums[low], nums[j] = nums[j], nums[low]
    return j

def quick_sort(nums, low, high):
    if low < high :
        p_index = partition(nums, low, high)
        quick_sort(nums, low, p_index - 1)
        quick_sort(nums, p_index+1, high)


numbers = [89,49,34,24,48,12,56]
quick_sort(numbers, 0, len(numbers) - 1)
print("Sorted array:", numbers)
```
